## Part 2 - Lab
Now that you know how to use Matplotlib, it is time to get back to helping the marketing manager. In the folder "data" you can find the csv file we generated yesterday. As mentioned before, the marketing manager is interested in knowing which sports have gotten most publicity in the past few years and hence don't need more publicity and which ones have not gotten enough publicity. There is two visualisation we would like you to produce:
    1) a bar chart that allows the marketing manager to compare the mentions of the featured sports
    2) a line chart that tracks the change in mentions of the top 3 mentioned sports over the past three years

```python
import matplotlib.pyplot as plt
import pandas as pd

newData = pd.read_csv("data/article_previews.csv")
newData
```

```python
sportList = newData["tag"]

sportDict = {}

for sport in sportList:
    if sport in sportDict:
        sportDict[sport] += 1
    else:
        sportDict[sport] = 1

        
sportKeys = list(sportDict.keys())

for sport in sportKeys:
    if sportDict[sport] < 10:
        sportDict.pop(sport)

sportNames = list(sportDict.keys())
sportMentions = list(sportDict.values())

plt.figure(figsize=(22,5))
plt.bar(sportNames, sportMentions)
```

```python
yearList = newData["year"]

basketballDict = {}
footballDict = {}
iceHockeyDict = {}

for i in range(len(sportList) -1, -1, -1):
    if sportList[i] == "Men's Basketball":
        if yearList[i] in basketballDict:
            basketballDict[yearList[i]] += 1
        else:
            basketballDict[yearList[i]] = 1
    elif sportList[i] == "Football":
        if yearList[i] in footballDict:
            footballDict[yearList[i]] += 1
        else:
            footballDict[yearList[i]] = 1
    elif sportList[i] == "Men's Ice Hockey":
        if yearList[i] in iceHockeyDict:
            iceHockeyDict[yearList[i]] += 1
        else:
            iceHockeyDict[yearList[i]] = 1

basketball = list(basketballDict.values())
football = list(footballDict.values())
iceHockey = list(iceHockeyDict.values())

football.pop()
iceHockey.pop()

years = [2019, 2020, 2021]

plt.plot(years, basketball, "g", label="Basketball")
plt.plot(years, football, "y", label="Football")     
plt.plot(years, iceHockey, "r", label="Ice Hockey")  

plt.xlabel('Year')
plt.ylabel('Mentions')

plt.legend(loc="upper center")

plt.show()
```
