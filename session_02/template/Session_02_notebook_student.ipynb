## Task #2 - generate a file with information for the marketing manager

OK, now that you know how to retrieve this information, write a program that extracts this information and writes it into the csv file "article_previews.csv". If you can't quite remember how you create and write into files, take a look at yesterday's assignment.

```python
import re
import requests
import csv
from bs4 import BeautifulSoup

url = "https://www.thecrimson.com/tag/sports-features/page/1/"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
soup.prettify()

articles = soup.find_all("div", class_="column-item")

with open("articleInfo.csv", "w") as file_write:
    writer = csv.writer(file_write)
    for article in articles:
        tag = article.find("div", class_="subsection css-t1s1ya-Fa")
        title = article.find("h1")
        weirdDate = article.find("div", class_="image-container")
        weirdList = re.findall("[0-9]+", str(weirdDate))
        year = weirdList[0]   
        writer.writerow([tag.text, title.text, year])
```

## Bonus Tasks
Congratulations! You have created a csv file, which we can use to explore our extracted data. So far, the file, however, only features the most recent articles, i.e. the articles from page 1 of the Sport's Features section. Write a program that creates a csv file with tag, title, and year of all articles from 2021-2019. 


```python
import requests
import csv
from bs4 import BeautifulSoup

def scrapeArticle(url):
    page = requests.get(url)

    soup = BeautifulSoup(page.content, "html.parser")
    soup.prettify()
    
    articles = soup.find_all("div", class_="column-item")

    infoList = []
    
    for article in articles:
        tag = article.find("div", class_="subsection css-t1s1ya-Fa")
        title = article.find("h1")
        weirdDate = article.find("div", class_="image-container")
        weirdList = re.findall("[\d]{4}", str(weirdDate))
        year = weirdList[0]   
        infoList.append([tag.text, title.text, year])
    
    return infoList

with open("articleInfo.csv", "w") as file_write:
    writer = csv.writer(file_write)
    writer.writerow(["tag", "title", "year"])
    infoList = []
    for i in range(1, 50):
        infoList += scrapeArticle("https://www.thecrimson.com/tag/sports-features/page/" + str(i) + "/")
    for i, row in enumerate(infoList):
        if int(row[2]) < 2019:
            break
        writer.writerow(row)
    
print("done")
```
